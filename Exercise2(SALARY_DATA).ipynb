```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Danxi3/GuzmanDaniella_BSIS2/main/SALARY_DATA%20(1).csv')
```

```python
print(df)
```

```python
salary = df['Salary'].to_numpy()
```

```python
mean_salary = np.mean(salary)
```

```python

print(mean_salary)
```

```python
median_salary = np.median(salary)
```

```python
print(median_salary)
```

```python
std_salary = np.std(salary)
```

```python
print(std_salary)
```

```python
total_salary = np.sum(salary)
```

```python
print(total_salary)
```

```python
age_info = df[df['Age'].notnull()][['Age']]

#Dropping values of age = 0
age_info_clean = age_info[age_info['Age'] != 0]

age_info_clean.info()
```

```python
fig = plt.figure(figsize = (20,8))
Age_plot = sns.distplot(age_info_clean, bins = 30)
Age_plot.tick_params(labelsize = 13)
std = age_info['Age'].std()

plt.xlabel('Age', fontdict = {'fontsize':15})
plt.title('Age of disease-affected humans', fontdict = {'fontsize':15})
plt.bold=True
plt.axvline(std,color = 'red')
plt.show()
```
